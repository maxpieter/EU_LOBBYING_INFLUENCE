## Word Clouds: Google vs Volkswagen

Filter all commission meetings by `organizations_raw` for each org, then build word clouds from `points_raised` + `conclusions`.

```python
# Fetch ALL commission meetings (paginated)
PAGE_SIZE = 1000
all_rows = []
offset = 0

while True:
    batch = (
        client.table("commission_meetings")
        .select("*")
        .range(offset, offset + PAGE_SIZE - 1)
        .execute()
    )
    rows = batch.data or []
    all_rows.extend(rows)
    if len(rows) < PAGE_SIZE:
        break
    offset += PAGE_SIZE

df_all = pd.DataFrame(all_rows)
print(f"Total meetings fetched: {len(df_all)}")

# Filter for Google and Volkswagen
df_google = df_all[df_all["organizations_raw"].str.contains("Google", case=False, na=False)].copy()
df_volkswagen = df_all[df_all["organizations_raw"].str.contains("Volkswagen", case=False, na=False)].copy()

print(f"Google meetings:     {len(df_google)}")
print(f"Volkswagen meetings: {len(df_volkswagen)}")
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
import re

# Standard English stop words + domain boilerplate
STOP_WORDS = {
    # common English
    "the", "and", "for", "that", "this", "with", "from", "are", "was", "were",
    "been", "being", "have", "has", "had", "will", "would", "could", "should",
    "can", "may", "might", "shall", "not", "but", "also", "its", "their",
    "they", "them", "than", "then", "into", "about", "which", "when", "where",
    "who", "whom", "how", "what", "there", "here", "these", "those", "other",
    "more", "most", "some", "any", "all", "each", "every", "both", "such",
    "only", "very", "too", "own", "over", "under", "between", "through",
    "during", "before", "after", "above", "below", "out", "off", "does",
    "did", "doing", "done", "just", "because", "while", "once", "until",
    "further", "same", "well", "back", "even", "still", "new", "now", "way",
    # short words / pronouns
    "a", "an", "of", "to", "in", "on", "at", "by", "is", "it", "or", "as",
    "if", "so", "no", "up", "do", "be", "we", "he", "she", "his", "her",
    # meeting boilerplate
    "commission", "representatives", "representative", "introduced", "company",
    "companies", "took", "note", "cab", "meeting", "met", "discussed",
    "raised", "points", "conclusions", "regarding", "noted", "presented",
    "position", "outlined", "reported", "views", "expressed", "stressed",
    "highlighted", "emphasised", "emphasized", "welcomed", "called", "urged",
    "stated", "indicated", "mentioned", "explained", "provided", "requested",
    "supported", "proposed", "agreed", "acknowledged", "commissioner", "commissioners", "google", "volkswagen", "aktiengesellschaft", "ag", "vw", "alphabet", "inc", "lobbying", "influence", "eu", "plan", "including"
}

def build_word_freq(df_org):
    """Combine points_raised + conclusions into word frequencies."""
    texts = []
    for _, row in df_org.iterrows():
        for col in ["points_raised", "conclusions"]:
            val = row.get(col)
            if pd.notna(val) and str(val).strip().lower() not in {"n/a", "na", "none", "nan", ""}:
                texts.append(str(val))
    
    combined = " ".join(texts).lower()
    # Keep only letters and spaces
    combined = re.sub(r"[^a-z\s]", "", combined)
    words = combined.split()
    # Filter stop words and very short words
    words = [w for w in words if w not in STOP_WORDS and len(w) > 2]
    return Counter(words)

google_freq = build_word_freq(df_google)
vw_freq = build_word_freq(df_volkswagen)

print(f"Google — unique words: {len(google_freq)}, total words: {sum(google_freq.values())}")
print(f"Volkswagen — unique words: {len(vw_freq)}, total words: {sum(vw_freq.values())}")
print(f"\nGoogle top 15:     {google_freq.most_common(15)}")
print(f"Volkswagen top 15: {vw_freq.most_common(15)}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

for ax, freq, name, color in [
    (axes[0], google_freq, "Google", "Blues"),
    (axes[1], vw_freq, "Volkswagen", "Reds"),
]:
    wc = WordCloud(
        width=900,
        height=500,
        background_color="white",
        colormap=color,
        max_words=100,
        prefer_horizontal=0.7,
        min_font_size=8,
    ).generate_from_frequencies(freq)
    
    ax.imshow(wc, interpolation="bilinear")
    ax.set_title(f"{name} — Commission Meetings", fontsize=18, fontweight="bold", pad=12)
    ax.axis("off")

plt.tight_layout()
plt.show()
```
